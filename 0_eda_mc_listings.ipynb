```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('data/mc_listings_extract.csv.gz', compression='gzip'
                 ,dtype={'scraped_at':str, 'status_date':str, 'first_scraped_at':str})
print(df.shape)
```

```python
# fill missing dates with 0 (1969)
df[['scraped_at','status_date','first_scraped_at']] = df[['scraped_at','status_date','first_scraped_at']].fillna(0).astype(int)
print(datetime.datetime.fromtimestamp(0))
```

```python
df['scraped_at'] = df['scraped_at'].map(lambda x: datetime.datetime.fromtimestamp(int(x)))
df['status_date'] = df['status_date'].map(lambda x: datetime.datetime.fromtimestamp(int(x)))
df['first_scraped_at'] = df['first_scraped_at'].map(lambda x: datetime.datetime.fromtimestamp(int(x)))
```
